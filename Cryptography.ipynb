 ### Cesar Cipher:
  It is a simple cipher in which each letter is replaced by a letter that is a certain number of letters away(shift). 

  The following python function takes a string(to be decrypted) and shift as input and applies shift to each letter. 

  The function will only accept and return lower case letters and spaces will not be changed (Does not verify this condition!)

```python
def cesar_crypt(text, shift):
  result = ' '
  for letter in text:
    if letter== ' ':
      result = result + ' '
    else:
      number= ord(letter)-97
      number = (number + shift)%26
      result = result + chr(number + 97)
  return result
```

```python
string1 = 'pbatenghyngvbaf lbh unir fhpprrqrq va qrpelcgvat gur fgevat'
```

```python
cesar_crypt(string1, 13)
```

```python
string2 = 'gwc uivioml bw nqvl bpm zqopb apqnb'
```

```python
shift =0
while shift >= -25:
  print(cesar_crypt(string2, shift))
  shift = shift-1
```

Clearly, shift =8!
